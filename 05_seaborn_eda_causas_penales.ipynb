```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
```

```python
#warnings.filterwarnings(action="ignore")
#sns.set_theme(style="whitegrid")
```

```python
data_path = Path("data")
```

```python
df_2017 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2017.csv", low_memory=False)
df_2018 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2018.csv", low_memory=False)
df_2019 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2019.csv", low_memory=False)
df_2020 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2020.csv", low_memory=False)
df_2021 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2021.csv", low_memory=False)
df_2022 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2022.csv", low_memory=False)
df_2023 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2023.csv", low_memory=False)
df_2024 = pd.read_csv(data_path / "poderes-judiciales-causas-penales-2024.csv", low_memory=False)
```

```python
df = pd.concat(
                [df_2017,
                df_2018,
                df_2019,
                df_2020,
                df_2021,
                df_2022,
                df_2023,
                df_2024],
                ignore_index = True)
```

```python
# Filtrado de columnas de interes
cols = [
    "causa_fecha_hecho",
    "caso_fecha_inicio",
    "fecha_envio",
    "provincia_nombre",
    "delito_descripcion",
    "delito_estadistico",
]

df_filtrado = df[cols].copy()
```

```python
df_filtrado
```

```python
df_filtrado["caso_fecha_inicio"] = df_filtrado["caso_fecha_inicio"].astype("str")
df_filtrado["causa_fecha_hecho"] = df_filtrado["causa_fecha_hecho"].astype("str")
```

```python
df_filtrado.info()
```

```python
df_filtrado["anio"] = df_filtrado["caso_fecha_inicio"].str[:4].astype(int)
```

```python
df_filtrado["anio"].isna().sum()
```

```python
df_filtrado
```

```python
causas_por_anio = (
                    df_filtrado
                    .groupby(["anio", "provincia_nombre"])
                    .size()
                    .reset_index(name="cantidad_causas")
                    )

causas_por_anio
```

```python
causas_por_delito = (
                    df_filtrado
                    .groupby(["anio","provincia_nombre","delito_estadistico"])
                    .size()
                    .reset_index(name="cantidad_causas")
                    )

causas_por_delito
```

1️⃣ Análisis temporal base

Pregunta: ¿cómo evoluciona la cantidad de causas en el tiempo?

```python
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=causas_por_anio,
    x="anio",
    y="cantidad_causas",
    estimator="sum",
    errorbar= None
)
plt.title("Cantidad total de causas penales por año")
plt.ylabel("Cantidad de causas")
plt.xlabel("Año")
plt.show()
```

2️⃣ Comparación entre provincias

Pregunta: ¿todas se comportan igual o hay sistemas judiciales con personalidad propia?

```python
g = sns.FacetGrid(
    causas_por_anio,
    col="provincia_nombre",
    col_wrap=3,
    height=3,
    sharey=False
)

g.map_dataframe(
    sns.lineplot,
    x="anio",
    y="cantidad_causas"
)

g.set_titles("{col_name}")
g.set_axis_labels("Año", "Cantidad de causas")
plt.show()
```

3️⃣ Delitos a lo largo del tiempo

Pregunta: ¿qué tipos de delitos explican los cambios?

Primero filtramos a los más frecuentes

```python
top_delitos = (
                causas_por_delito
                .groupby("delito_estadistico")["cantidad_causas"]
                .sum()
                .sort_values(ascending=False)
                .head(10)
                .index
                )

top_delitos
```

```python
df_top = causas_por_delito[
                            causas_por_delito["delito_estadistico"].isin(top_delitos)
                            ]
df_top
```

```python
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=df_top,
    x="anio",
    y="cantidad_causas",
    hue="delito_estadistico",
    #hue="provincia_nombre",
    estimator="sum",
    errorbar= None
)

plt.title("Evolución temporal de los principales tipos de delito")
plt.ylabel("Cantidad de causas")
plt.xlabel("Año")
plt.legend(title="Delito")
plt.show()
```

```python
g = sns.FacetGrid(
    df_top,
    col="provincia_nombre",
    col_wrap=3,
    height=3,
    sharey=False
)

g.map_dataframe(
    sns.lineplot,
    x="anio",
    y="cantidad_causas"
)

g.set_titles("{col_name}")
g.set_axis_labels("Año", "Cantidad de causas")
plt.show()
```

4️⃣ Comparaciones avanzadas (Seaborn puro)
Boxplot

Distribución de causas por delito a lo largo de los años

```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_top,
    x="delito_estadistico",
    y="cantidad_causas"
)
plt.xticks(rotation=45)
plt.title("Distribución de causas por tipo de delito")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(
    data=df_top,
    x="delito_estadistico",
    y="cantidad_causas",
    estimator="mean",
    ci=None
)
plt.xticks(rotation=45)
plt.title("Promedio anual de causas por tipo de delito")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.pointplot(
    data=df_top,
    x="anio",
    y="cantidad_causas",
    hue="delito_estadistico"
)
plt.title("Evolución con intervalos de confianza")
plt.show()
```
