### Análisis simple pero potente

```python
df
```

```python
# Segun el Sexo, dame el promedio de sobrevivientes de cada tipo
df.groupby("Sex")["Survived"].mean()
```

```python
# Segun el la clase social, dame el promedio de sobrevivientes de cada tipo
df.groupby("Pclass")["Survived"].mean()
```

### ordenar resultados

```python
# Ordena valores por precio del boleto del mas caro al mas barato
df.sort_values("Fare", ascending=False).head()
```

## Conclusiones

- La supervivencia varía fuertemente según sexo y clase.
